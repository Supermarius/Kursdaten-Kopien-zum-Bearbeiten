## Explorieren

Ihr Ziel ist, aus den Börsendaten eine einfache Story zu generieren.

Überlegen Sie sich dazu:
- Was sind mögliche interessante Fragen?
- Welche Auswertungen sind nötig, um die Fragen zu beanworten?
- Wie stellen Sie die Resultate mittels Grafiken dar?

Arbeiten Sie in Gruppen, um ein Ergebnis zu erarbeiten!

```python
df
```

```python
df_change = df.resample("MS").mean().pct_change()
df_change
```

```python
df_volatil = df_change[(df_change >= 0.05)|(df_change <= -0.05)]
```

```python
df_volatil
```

```python
df_volatil.count().sort_values(ascending=False).plot(kind="bar", figsize=(10,5))
```

```python
performance = df.iloc[-1]/df.iloc[-0]
performance
```

```python
volatile_monate = df_volatil.count()
volatile_monate
```

```python
relativiert = volatile_monate/performance
relativiert.sort_values(ascending = False).plot(kind="bar", figsize=(10,5), title = "Risikoanalyse: Volatilität, Performance-bereinigt (negative Ausschläge zählen mehr)")
```

```python
# neuer Versuch: positive und negative Werte im gleichen Bar-Chart:
```

```python
df_volatil_positiv = df_change[df_change >= 0.05]
df_volatil_negativ = df_change[df_change <= -0.05]
```

```python
monate_negativ = (df_volatil_negativ.count()) * -1
```

```python
monate_positiv = df_volatil_positiv.count()
```

```python
#Jetzt verwandeln wir die Pandas-Series in drei Listen: den Index und die Anzahl poitiver und negativer Monate.
liste_firmen = monate_positiv.index.tolist()
liste_positiv = monate_positiv.tolist()
liste_negativ = monate_negativ.tolist()
```

```python
#daraus machen wir ein Dictionary
dictionary = {"Firmen": liste_firmen, "+": liste_positiv, "-": liste_negativ}
```

```python
#und daraus wieder eine Dataframe
df = pd.DataFrame(dictionary)
```

```python
df.plot(kind="bar", x="Firmen", figsize=(15,10) )
```

```python
#Jetzt wollen wir das noch sauber sortieren. Dazu zuerst in einer neuen Spalte die Bandbreite bestimmen:
df["Bandbreite"] = df["+"]-df["-"]
df.head(5)
```

```python
df.sort_values("Bandbreite", ascending = False).plot(kind="bar", x="Firmen", y= ["+", "-"], figsize=(15,10), grid =1 )
```
