#### Übung: 3 Tage aus Erdbeben-DB abfragen

```python
url_anfang = "https://earthquake.usgs.gov/fdsnws/event/1/count?starttime=2019-09-0"
url_mitte = "&endtime=2019-09-0"

# einen loop einbauen mit einer Variabel (dates), die wir gar nicht brauchen, und einer Durchlaufzahl (range):

for day in range(1,4):
    r = requests.get(url_anfang + str(day) + url_mitte + str(day+1))
    day=day+1
    print("2019-09-0" + str(day) + " bis 2019-09-0" + str(day+1) + ": "+ (r.text))
```

```python
# Variante mit Speichern der resultate in einer Liste statt print:

url_anfang = "https://earthquake.usgs.gov/fdsnws/event/1/count?starttime=2019-09-0"
url_mitte = "&endtime=2019-09-0"

# einen loop einbauen mit einer Variabel (dates), die wir gar nicht brauchen, und einer Durchlaufzahl (range):

result_list = []

for day in range(1,4):
    abfrage = requests.get(url_anfang + str(day) + url_mitte + str(day+1))
    day=day+1
    result_list.append(abfrage.text)

for line in result_list:
    print(line)
```

```python
# Mit Datumsfunktionen arbeiten:

import datetime
base = datetime.datetime.today()
date_list = [base - datetime.timedelta(days=x) for x in range(0,10)]
for date in date_list:
    print(date.strftime("%Y-%m-%d"))
    
# > könnte man jetzt in Abfrageloop einbauen
```

```python
query = requests.get("https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2019-09-18&endtime=2019-09-23")
```

```python
# Als json importieren (als Text wäre es ein langer Stringund nicht zu verarbeiten...)
query = query.json()
```

```python
# Jetzt geht es darum, dieses Dictionary zu untersuchen:
len(query)
```

```python
# Hier fragen wir die Keys ab:
for key in query:
    print(key)
```

```python
erdbeben = query["features"]
erdbeben
# Das Ergebnis ist eine Liste voller Dictionarys:
```

```python
erdbeben[0]
```

```python
erdbeben[0]["properties"]["place"]
```

```python
Beben_Dict = {"Ort": erdbeben[0]["properties"]["place"], "Stärke":erdbeben[0]["properties"]["mag"]}
```

```python
leereliste = []

for einzelbeben in erdbeben:
    place = (einzelbeben["properties"]["place"])
    magnitude = (einzelbeben["properties"]["mag"])
    
    Beben_Dict = {"Ort":place, "Stärke":magnitude}
    # (Dieser Dictionary wird bei jedem Zyklus mit den neuen Angaben überschrieben)
    
    leereliste.append(Beben_Dict)
    # (Die Dict-Daten jedes Zyklus sammeln wir in einer Liste)

#Pandas-Funktion zum schönen Anzeigen vonListe
pd.DataFrame(leereliste) 
```

```python
# Das Ganze filtern, so dass nur die grosen Beben (> 5) angezeigt werden

leereliste = []

for einzelbeben in erdbeben:
    if (einzelbeben["properties"]["mag"]) > 5:
        place = (einzelbeben["properties"]["place"])
        magnitude = (einzelbeben["properties"]["mag"])
    
        Beben_Dict = {"Ort":place, "Stärke":magnitude}
        # (Dieser Dictionary wird bei jedem Zyklus mit den neuen Angaben überschrieben)
    
        leereliste.append(Beben_Dict)
        # (Die Dict-Daten jedes Zyklus sammeln wir in einer Liste)

#Pandas-Funktion zum schönen Anzeigen von Liste
pd.DataFrame(leereliste) 
```

### Speichern des Ganzen in einem leeren csv-file

```python
pd.DataFrame(leereliste).to_csv("Meine_Liste")
```
