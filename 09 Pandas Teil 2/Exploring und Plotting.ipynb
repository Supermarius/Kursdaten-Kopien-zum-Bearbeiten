### Neues Experiment: Jene Dez bis März summieren, die Durchschnittstemp. > 0 hatten

```python
df= pd.read_csv(path, delimiter = "\t", skiprows = 15)

# zuerst füge ich der Tabelle Trze/False Kolumnen hinzu
df["Dez_nullplus"] = df["dec"] > 0
df["Jan_nullplus"] = df["jan"] > 0
df["Feb_nullplus"] = df["feb"] > 0
df["Mär_nullplus"] = df["mar"] > 0
```

```python
# jetzt mache ich, dass alle False in O (und True in 1) umgewandelt werden:
df["Dez_nullplus"] = df["Dez_nullplus"].replace(False, 0)
df["Jan_nullplus"] = df["Jan_nullplus"].replace(False, 0)
df["Feb_nullplus"] = df["Feb_nullplus"].replace(False, 0)
df["Mär_nullplus"] = df["Mär_nullplus"].replace(False, 0)
```

```python
# hier stelle ich ein, dass ich alle Kolumnen sehe:
pd.set_option("display.max_columns", 50)
# dann mache ich einen neuen Frame (copy!), der nur meine 4 Kolumnen plus das Jahr enthält
df_neu = df[["time","Dez_nullplus","Jan_nullplus", "Feb_nullplus", "Mär_nullplus"]].copy()
df_neu
```

```python
df_neu["summe"] = df_neu["Dez_nullplus"] + df_neu["Jan_nullplus"] + df_neu["Feb_nullplus"] + df_neu["Mär_nullplus"]
df_neu[["summe", "time"]].plot(kind = "scatter", x="time", y="summe", title = "Wintermonate mit mehr als 0 Grad in Schnitt")
```
